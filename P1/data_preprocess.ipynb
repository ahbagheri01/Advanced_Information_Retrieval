````python
#file 1 & 2
import re
import pandas as pd
import json

data = {}
dup = []
for ind in [1,2]:
    f1 = open(f"data/vehicles{ind}.txt","r",encoding = "utf-8")
    f1.readline()
    for line in f1:
        l = line.strip()
        if l in data :
            dup.append(l)
            continue
        if len(l) > 1:
            data[l] = 1
    f1.close()
print(len(data))
f3 = open("data/vehicles3.txt","r",encoding = "utf-8")
f3.readline()
name = ""
for line in f3:
    l = line.strip()
    ll = l.split("-")
    try:
        if len(ll) == 2 and int(ll[0]) and int(ll[1]):
            if not name.strip() in data:
                data[name.strip()] = 1
            name = ""
            continue
    except:
        pass
    name += l 
print(len(data))
for ind in [4]:
    f1 = open(f"data/vehicles{ind}.txt","r",encoding = "utf-8")
    f1.readline()
    for line in f1:
        l = line.strip()
        if l in data :
            dup.append(l)
            continue
        if len(l) > 1:
            data[l] = 0
    f1.close()
add = ["دنده ای","دنده‌ای","اتومات","دنده اتومات","دنده‌اتومات"]
before = ["خودرو","خودروی"]
new_d = {}
remove_list = ["هن",
"بی",
"دی",
"جی",
"ام",
"گک",
"سه",
"مگ"]
for car in data:
    print(car) if len(car) == 2 else None 
    car = car.replace("‌"," ")
    car = car.strip()
    cars = car.split(" ")
    size = len(cars)
    if cars in remove_list:
        continue
    for i in range(1,size+1):
        new_d[" ".join(cars[:i])] = " ".join(cars[:i])
        new_d["‌".join(cars[:i])] = "‌".join(cars[:i])
        for additional in add:
            if data[car]:
                new_d[" ".join(cars[:i])+" " +additional] = " ".join(cars[:i])+" " +additional
                new_d["‌".join(cars[:i])+"‌" +additional] = "‌".join(cars[:i])+"‌" +additional
        for bef in before:
            if data[car]:
                new_d[bef+" "+" ".join(cars[:i])] = bef+" ".join(cars[:i])
                new_d[bef+"‌"+"‌".join(cars[:i])] = bef+"‌".join(cars[:i])

d = json.load(open("data/vehicle_type.json","r",encoding="utf-8"))
for i in d:
    l = i.strip()
    if l in new_d:
        continue
    new_d[l] = l
print(len(new_d))

def checkInt(str):
    try:
        int(str)
        return False
    except ValueError:
        return True
d = pd.DataFrame({"vehicles":[key.strip() for key in new_d if checkInt(key.strip())]})
d.to_csv("resources/vehicles.csv",index = False)
print(len(d))
````

````python
import json
import re
places = []
f  = open(f"data/iran_cities_better.txt","r",encoding = "utf-8")
i = 0
j = 0
for line in f:
    i+=1
    l = line.strip()
    l = re.findall("[^\(|^\)]+", l)[0].strip()
    if l in places:
        j+=1
        continue
    if len(l) > 1:
        places.append(l)
        pass
f.close()
print(i,j)
pl = ["provinces","countries","cities","special_place"]
for p in pl:
    new_files = json.load(open("data/provinces.json","r",encoding="utf-8"))
    for place in new_files:
        l = place.strip()
        if l in places:
            continue
        if len(l) > 1:
            places.append(l)
d = pd.DataFrame({"place":places})
d.to_csv("resources/places.csv",index = False)
````
